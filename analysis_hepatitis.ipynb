
## Análise Exploratória do Hepatitis Dataset



## Introdução

Este projeto tem como objetivo realizar uma análise exploratória de dados (EDA)
sobre um conjunto de dados clínicos de pacientes com hepatite, utilizando Python
em ambiente Jupyter Notebook.


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```


## Carregamento e Descrição do Dataset


```python
import pandas as pd

df = pd.read_csv("hepatitis_dataset.csv")

df.head()
```

```python

df.info()
```

```python

df.describe(include="all")
```


## Limpeza e Preparação dos Dados


```python

# Verificar valores em falta
df.isnull().sum()
```

```python

# Separar variáveis numéricas e categóricas
num_cols = df.select_dtypes(include=['int64','float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

# Imputação
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python

# Codificação de variáveis binárias
binary_map = {
    'yes': 1, 'no': 0,
    'male': 1, 'female': 0,
    'LIVE': 1, 'DIE': 0
}

df.replace(binary_map, inplace=True)
```

```python

# Criação de faixas etárias
bins = [0, 30, 50, 70, 100]
labels = ['Jovem', 'Adulto', 'Meia-idade', 'Idoso']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```


## Análise Exploratória (EDA)


```python

# Pergunta 1: Taxa de sobrevivência
df['class'].value_counts(normalize=True) * 100
```

```python

df['class'].value_counts().plot(kind='bar')
plt.title('Distribuição do Desfecho Clínico')
plt.xlabel('Classe (1 = LIVE, 0 = DIE)')
plt.ylabel('Número de Pacientes')
plt.show()
```

```python

# Pergunta 2: Idade vs desfecho
df.groupby('class')['age'].describe()
```

```python

sns.boxplot(x='class', y='age', data=df)
plt.title('Idade por Desfecho Clínico')
plt.show()
```

```python

# Pergunta 3: Bilirrubina vs desfecho
df.groupby('class')['bilirubin'].mean()
```

```python

sns.boxplot(x='class', y='bilirubin', data=df)
plt.title('Bilirrubina por Desfecho Clínico')
plt.show()
```

```python

# Pergunta 4: Albumina vs desfecho
df.groupby('class')['albumin'].mean()
```

```python

sns.boxplot(x='class', y='albumin', data=df)
plt.title('Albumina por Desfecho Clínico')
plt.show()
```

```python

# Pergunta 5: Ascite vs mortalidade
pd.crosstab(df['ascites'], df['class'], normalize='index') * 100
```

```python

pd.crosstab(df['ascites'], df['class']).plot(kind='bar', stacked=True)
plt.title('Ascite vs Desfecho Clínico')
plt.show()
```

```python

# Correlação entre variáveis laboratoriais
corr = df[['age','bilirubin','alk_phosphate','sgot','albumin','protime']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlação entre Variáveis Laboratoriais')
plt.show()
```


## Conclusões

O dataset encontra-se limpo e preparado para análise exploratória. A análise
demonstrou que idade, indicadores laboratoriais e sintomas clínicos graves
estão fortemente associados ao desfecho clínico da hepatite.
